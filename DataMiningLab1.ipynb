Before beginning with question one, I checked which columns were string and which were numeric so I knew which columns needed transforming before producing the five-number summaries. After discovering that even MonthlyCharges, which appeared numeric, was string, it was transformed to a float. The variable Churn was recoded to the variable churn as numeric as well. Tenure was already an integer so no transformation was needed. Creating a boxplot for the class attribute MonthlyCharges showed that there were no outliers in this category, but there was still quite a large range in monthly charges for Telco's list of customers. The five-number summaries of each of the selected variables showed that only around 25% of customers have left within the past month (churn summary), the most a customer pays per month at Telco is $118.75 (MonthlyCharges summary), and that the customer with the longest tenure has shopped at Telco for 72 years! When examining the euclidean distance, it appears that tenure is more similar to the class attribute MonthlyCharges. These results could be useful for examining the relationship between customers tenure / monthly charges and their impact on churn. I could imagine a company performing a regression to see whether or not these two are good predictors for churn to gain insight on why customers are leaving Telco for a rival company.
